```python
from ultralytics import YOLO

# Load a pre-trained model (you can use yolov8s.pt, yolov8m.pt, etc.)
# yolo_list = ["yolov8n.pt", "yolov8s.pt", "yolo11n.pt", "yolo11s.pt"]
yolo_list = ["yolo11s.pt"]
conf_yaml = "/Volumes/Cucumber/Projects/datasets/process/2025-07-18/sku110k-only-yolo/custom_dataset.yaml"
for yolo_pt in yolo_list:
    model = YOLO(yolo_pt)
    model.train(
        data=conf_yaml,  # path to the dataset config file
        epochs=32,
        imgsz=640,
        batch=4,
        name=yolo_pt,
        augment=True
    )
    metrics = model.val(
        data=conf_yaml,
        split="val"  # or "val" for validation
    )
    print("Validation", metrics.results_dict)

    metrics = model.val(
        data=conf_yaml,
        split="test"  # or "val" for validation
    )
    print("Test", metrics.results_dict)
```
