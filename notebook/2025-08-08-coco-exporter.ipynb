```python
def get_dataset_by_split_name(split_name: str):
    lsb_connector = LabelStudioConnector(
        url=appconfig.LABEL_STUDIO_URL,
        api_key=appconfig.LABEL_STUDIO_API_KEY,
        project_id=appconfig.LABEL_STUDIO_PROJECT_MAPPING[split_name],
        temp_dir=appconfig.LABEL_STUDIO_TEMP_DIR
    )
    tasks = lsb_connector.get_tasks()
    dataset = lsb_connector.download_dataset(tasks, LABEL_2_IDX)
    return dataset
```

```python
train = get_dataset_by_split_name("train")
val_test = get_dataset_by_split_name("val")
val_test.extend(get_dataset_by_split_name("test"))
```

### Modify metadata to match SKU110k

```python
split_name = "val"
with open(os.path.join(path_merge, f"{split_name}/annotations_{split_name}.json")) as f:
    d = json.load(f)
    id = 1
    d["categories"] = [
        {
            "id": id,
            "name": "object",
            "supercategory": "retail_item"
        }
    ]

    for a in d["annotations"]:
        a["category_id"] = id

with open(os.path.join(path_merge, f"{split_name}/annotations_{split_name}.json"), "w") as f:
    json.dump(d, f)
```
